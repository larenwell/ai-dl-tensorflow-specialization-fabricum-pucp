```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,3,4,7,11,15]
y = [10,20,25,35,40,60]

plt.plot(x,y, marker='o')
plt.title('Datos aleatorios')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
categorias = ['A','B','C','D','E']
valores = [24,12,23,22,25]

plt.bar(categorias, valores, color = ['red','blue','orange','purple','gray'])
plt.title('Grafico de barras')
plt.show()
```

```python
labels = ["Python", "Java", "C++"]
sizes = [50, 30, 20]

plt.pie(sizes, labels=labels)
plt.title("Lenguajes de programacion")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.title("Grafico de dispersion")
plt.show()
```

```python
data = np.random.randn(1000)
```

```python
plt.hist(data, bins=30)
plt.title("Distribucion")
plt.show()
```
