```python
import pandas as pd
```

Definición de datos

```python
data={'nombre':['Juan','Maria','Pedro','Ana'],
      'edad':[20,25,30,35],
      'ciudad':['Santiago','Valparaiso','Concepcion','Antofagasta']}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

Acceder a los datos

```python
# Acceder a los datos

# Se accede a los datos por medio de los nombres de las columnas, se muestra una serie
# Al menos que se especifique con doble corchete, se muestra un dataframe
print(df['nombre'])
print(df['edad'])
```

Empleando iloc (posición) - No se puede modificar

```python
df.iloc[0]
```

```python
df.iloc[0:2]
```

```python
df.iloc[:,0]
```

```python
# iloc — selección por posición entera (como índices de Python), 
# donde el slice es exclusivo en el final. 
# iloc[0:2] devuelve filas 0 y 1.
# Así mismo este criterio para columnas, para este caso se selecciona la columna 0

df.iloc[0:2,0]
```

```python
# Acá se seleccionan las filas 0 y 2 y la columna 0
df.iloc[[0,2],0]
```

```python
df
```

Empleando loc (Etiquetas) - Se puede modificar
Se modifica la etiqueta de los indices ()

```python
# loc — selección por etiquetas/nombres, 
# donde el slice es inclusivo en ambos extremos. 
# loc[0:2] devuelve filas 0, 1 y 2.

df.loc[0:2,'nombre']
```

```python
df.loc[df['edad'] > 24,'nombre']
```

Nulos

```python
df_null=pd.DataFrame({'categoria':['A','B','A','B','A','B'],
                      'ventas':[10,None,30,40,None,60],
                      'fecha':['2021-01-01','2021-01-02',None,'2021-01-04','2021-01-05','2021-01-06']})
```

```python
df_null
```

```python
df_null.isnull()
```

```python
df_null.isnull().sum()
```

```python
df_null
```

```python
# Eliminar filas que tengan AL MENOS un nulo
df_null.dropna()
```

```python
# Eliminar filas donde TODAS las columnas sean nulas
df_null.dropna(how='all')

```

```python
# Eliminar filas con nulos solo en columnas específicas
df_null.dropna(subset=['ventas'])
```

```python
# Eliminar COLUMNAS que tengan algún nulo
df_null.dropna(axis=1)
```

```python
df_null
# Llenar todos los nulos con un valor fijo
```

```python
# Llenar todos los nulos con un valor fijo
df_null.fillna(0)
```

```python
# Llenar con la MEDIA de cada columna
df_null.fillna(df_null.mean())

# Tener cuidado con el tipo de dato de la columna
#df_null.dtypes

# Llenar con la MEDIA de cada columna
```

```python
df_null['ventas'].fillna(df_null['ventas'].mean())
```

```python
# Llenar con valores distintos por columna
df_null.fillna({'ventas': df_null['ventas'].mean(), 'fecha': '2026-01-01'})
```

Group by

```python
df_group=pd.DataFrame({'categoria':['A','B','A','B','A','B'],
                      'ventas':[10,20,30,40,50,60]})
```

```python
df_group.groupby('categoria').sum()
```

```python
df_group.groupby('categoria').mean()
```

```python
df_group.groupby('categoria').count()
```
